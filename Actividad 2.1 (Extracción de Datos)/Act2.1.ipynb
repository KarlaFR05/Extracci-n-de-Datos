```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#CONVERTIR BASE DE DATOS A INFORMACIÓN EXTRAIBLE
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df.info()
```

```python
#FILTRO POR OBJETO, EN ESTE CASO SE SELECCIONA LA VARIABLE "NOMBRE_CLIENTE" PARA MOSTRAR LA INFORMACIÓN DEL CLIENTE JUAN ANTONIO RESENDIZ UGALDE
filtro1= df[(df["NOMBRE_CLIENTE"]) == "JUAN ANTONIO RESENDIZ UGALDE"]
filtro1
filtro1.to_csv("ACT2.1_FILTRO1.csv", index=False)
```

```python
#FILTRO POR FILAS
 #FILAS NO CONSECUTIVAS, EN ESTE SE SELECCIONARON LAS FILAS 2, 5, 15, 20, 22 PARA CONOCER INFORMACIÓN DE FORMA ALEATORIA
filtro2= df.iloc[[2, 5, 15, 20, 22], :]  
filtro2
filtro2.to_csv("ACT2.1_FILTRO2.csv", index=False)
```

```python
#FILTRO POR COLUMNAS 
#COLUMNAS NO CONSECUTIVAS, SE SELECCIONARON LAS COLUMNAS  3, 6, 7, 11 FILTAR LA INFORMACIÓN DEL NOMBRE DE LOS CLIENTES, LA CANTIDAD DE PREODUCTOS,
# DESCRIPCIÓN DEL PRODUCTO Y EL COSTO TOTAL
filtro3= df.iloc[: , [3, 4, 6, 11]] 
filtro3
filtro3.to_csv("ACT2.1_FILTRO3.csv", index=False)
```

```python
#CONVIERTE  A LA COLUMNA 3, EN ESTE CASO NOMBRE_CLIENTE EN EL INDICE DEL DATAFRAME
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=3)
df1
```

```python
#FILTRO FILAS VS COLUMNAS
#DEL DATAFRAME ANTERIOR SE SELECCIONAN LAS FILAS "ANA BARBARA CRUZ PALACIOS", "APLICACIONES SILIDRIEL" Y LA COLUMNA "MARGEN_TOTAL_CALCULADO" 
#Y COMO RESULTADO OBTENEMOS LA INFORMACIÓN QUE SE CONTIENE EN ESTAS FILAS CON RELACIÓN A LA COLUMNA 
filtro4= df1.loc[["ANA BARBARA CRUZ PALACIOS", "APLICACIONES SILIDRIEL"],["MARGEN_TOTAL_CALCULADO"]]
filtro4
filtro4.to_csv("ACT2.1_FILTRO4.csv", index=False)
```

```python
#FILTRO POR CABECERA 
#PROPORCIONA LA INFORMACIÓN DE LOS 5 PRIMEROS DATOS
filtro5= df.head()
filtro5
filtro5.to_csv("ACT2.1_FILTRO5.csv", index=False)
```

```python
#FILTRO COLA 
#PROPORCIONA LA INFORMACIÓN DE LOS 5 ULTIMOS DATOS
filtro6= df.tail()
filtro6
filtro6.to_csv("ACT2.1_FILTRO6.csv", index=False)
```

```python
#OBTERNER  LA INFORMACIÓN SOBRE  LAS COMPRAS QUE CUENTAN CON UN MARGEN TOTAL CALCULADO MAYOR A 30000
filtro7= df[df["MARGEN_TOTAL_CALCULADO"]> 30000]
filtro7
filtro7.to_csv("ACT2.1_FILTRO7.csv", index=False)
```

```python
#FILTRO Y
#MUESTRA LA INFORMACIÓN SOBRE QUIENES CUENTAN CON UN MARGEN TOTAL CALCULADO MAYOR  A 30000 Y SU FRECHA_DOC ES 2022-09-23 
#(CUMPLE CON AMBOS REQUISITOS)
filtro8= df[(df["MARGEN_TOTAL_CALCULADO"]> 30000) & (df["FECHA_DOC"]=="2022-09-23")]
filtro8
filtro8.to_csv("ACT2.1_FILTRO8.csv", index=False)
```

```python
#MUESTRA LA INFORMACIÓN SOBRE TODOS LOS QUE  CUENTAN CON UN MARGEN TOTAL CALCULADO MAYOR  A 30000, 
#O SU FRECHA_DOC ES 2022-09-23 (CUMPLE CON POR LO MENOS UN  REQUISITO)
filtro9= df[(df["MARGEN_TOTAL_CALCULADO"]> 30000) | (df["FECHA_DOC"]=="2022-09-23")]
filtro9
filtro9.to_csv("ACT2.1_FILTRO9.csv", index=False)
```

```python
#FILTRO NOT
#NOS PROPORCIONA NUESTRO DATAFRAME SIN MOSTRAR AL CLIENTE CON EL NOMBRE "CARCOUSTICS INDUSTRIAL DE MEXICO"
filtro10=df[~(df["NOMBRE_CLIENTE"]== "CARCOUSTICS INDUSTRIAL DE MEXICO")]
filtro10
filtro10.to_csv("ACT2.1_FILTRO10.csv", index=False)
```
