```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv('Valencia.csv', encoding='latin1')
df.info()
```

**Filtro 1 Resgistro de 5 "Host name" diferentes**

```python
filtro1 = df.tail()
filtro1
filtro1.to_csv("filtro1.csv")
```

**Filtro 2 Resgistro de host que se hallan unido a Airbnb "antes del año 2019"**

```python
df['host_since'] = pd.to_datetime(df['host_since'], errors='coerce')
filtro2 = df[df['host_since'] < pd.to_datetime('2019-01-01')]
filtro2
filtro2.to_csv("filtro2.csv")
```

**Filtro 3 Resgistro de host que responden  "a más tardar en 1 día"**

```python
filtro3= df[(df["host_response_time"]) < "within a day"]
filtro3
filtro3.to_csv("filtro3.csv")
```

**Filtro 4 los registros que tengan "al menos 5 accomodates"**

```python
filtro4= df[(df["accommodates"]) >= 5 ]
filtro4
filtro4.to_csv("filtro4.csv")
```

**Filtro 5 los registros de los cuartos: "Entire home" y "Private room"**

```python
filtro5= df[(df["room_type"]=="Entire home/apt") | (df["room_type"]=="Private room")]
filtro5
filtro5.to_csv("filtro5.csv")
```

**Filtro 6  los registros que "a lo mucho cuenten con 4 camas"**

```python
filtro6=  df[(df["bedrooms"] <= 4)]
filtro6
filtro6.to_csv("filtro6.csv")
```

**Filtro 7 Los registros que "tengan 1 baño o que sean superhost" pero con la condición de que "todos sean instant_booktable**

```python
filtro7= df[(df["bathrooms"] == 1) | (df["host_is_superhost"] == "t") & (df["instant_bookable"] == "t")]
filtro7
filtro7.to_csv("filtro7.csv")
```

**Filtro 8 "reviewew_scores_cleanliness" menor de 3 y "review_scores_rating" sea mayor de 3**

```python
filtro8=df[(df["review_scores_cleanliness"] <3 ) & (df["review_scores_rating"] >3)]
filtro8
filtro8.to_csv("filtro8.csv")
```
